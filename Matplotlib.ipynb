```python
import matplotlib.pyplot as plt
```

```python
Cantidad = [4976,4713,3890,3705,3653,3469]
etiquetas=['Mazda cx 30','Renault Kwid','Chevrolet Onix','Renault Stepway','Renault Duster', 'Renault Sandero']
colores=['red','orange','blue','green','white', 'yellow']
plt.pie(Cantidad,labels=etiquetas,colors=colores,wedgeprops={'edgecolor':'black'})
plt.title('Autos vendidos')
```

```python
import matplotlib.pyplot as plt
Cantidad = [4976,4713,3890,3705,3653,3469]
Precio = [108,44,52.7,50,74,48.8]
FuerzaC = [186,66,160,111,143,156]

colores = ["gold", "green", "blue","silver","white", "purple"]
plt.plot(Cantidad,Precio,color="green",linewidth=3,label="Precio (M)")
plt.plot(Cantidad,FuerzaC,color="red",linewidth=3,label="Caballos de Fuerza")
plt.legend()
plt.title("Autos vendidos \n VS \n Precio/Caballos de Fuerza")
plt.xlabel("Cantidad de autos vendidos.")
plt.ylabel("Caballos de fuerza/ Precio")
```
